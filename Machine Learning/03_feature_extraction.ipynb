```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

###Input data / Za≈Çadowanie danych

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df_raw = fetch_financial_data()
df_raw.head()
```

Backup data / Kopia danych

```python
df=df_raw.copy()
df.info()
```

Generate new variables / Generowanie nowych zmiennych

```python
df.index.month
```

```python
df['day']=df.index.day
df['month']=df.index.month
df['year']=df.index.year
df
```

Desctetization variable / Dyskretyzacja zmiennych

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=3)
df
```

```python
df['height_cat']=pd.cut(x=df.height, bins=(160,170,180,195))
df
```

```python
df['height_cat']=pd.cut(x=df.height, bins=(160,170,180,195), labels=['small', 'medium','heigh'])
df
```

```python
pd.get_dummies(df,drop_first=True, prefix='height')
```

###Feature extraction / Ekstrakcja cech

```python
df=pd.DataFrame(data={'lang':[['PL','ENG'],['GER','ENG','PL','FR'],['RUS']]})
df
```

```python
df['lang_number']=df['lang'].apply(len)
df
```

```python
df['PL_Flag']=df['lang'].apply(lambda x: 1 if 'PL' in x else 0)
df
```
