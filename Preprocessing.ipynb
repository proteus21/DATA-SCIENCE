```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

Imput data

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df_raw.info()
```

Backup data

```python
df=df_raw.copy()
df.info()
```

Change category of input data


```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col]=df[col].astype('category')

df['weight']=df['weight'].astype('float')
df.info()
```

```python
df.describe()
```

Transform

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
df.describe(include=['float64']).T
```

### LABEL ENCODING

Label encoding  always is applied to coding values as yes, no

```python
from  sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

To do faster

```python
le.transform(df['bought'])
```

```python
le.classes_
```

```python
df['bought']=le.fit_transform(df['bought'])
df
```

Get back to orginal input data

```python
le.inverse_transform(df['bought'])
```

```python
df['bought']=le.inverse_transform(df['bought'])
df
```

###STANDARIZATION - StandarScaler

```python
print(f"{df['price']} \n")
print(f"Srednia:{df['price'].mean()}")
print(f"Odchylenie standardowe:{df['price'].std():.3f}")
```

```python
(df['price']-df['price'].mean())/df['price'].std()
```

```python
def standarize(x):
  return (x-x.mean())/x.std()
```

```python
standarize(df['price'])
```

If we apply function scale, we simple calculate standarization but there is diffrence between above function  standarize(df['price']) and below 'Scale', because it is cause the fact
